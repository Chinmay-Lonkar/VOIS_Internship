```python
plt.figure(figsize=(10,6))
sns.barplot(x=df['Category'],y=df['Sales'],data=df)
```

```python
plt.figure(figsize=(10,6))
sns.barplot(x=df['Category'],y=df['Profit'],data=df)
```

```python
plt.figure(figsize=(10,6))
sns.barplot(x=df['Category'],y=df['Discount'],data=df)
```

```python
plt.figure(figsize=(8,5))
regional_sales = df.groupby('Region')['Sales'].sum().sort_values(ascending=False)
regional_sales.plot(kind='bar', xlabel='Regions', ylabel='Total Sales', title='Regional Sales Performance')
plt.show()
```

```python
plt.figure(figsize=(8,5))
segments = df.groupby('Segment')['Sales'].sum().sort_values(ascending=False)
segments.plot(kind='bar', xlabel='Segments', ylabel='Total Sales', title='Segment Analysis')
plt.show()
```

```python
plt.figure(figsize=(8,5))
city_data = df.groupby('City')['Sales'].sum().sort_values(ascending=False).head(10)
city_data.plot(kind='bar', xlabel='City', ylabel='Total Sales', title='Sales Ranking of the Top 10 Cities')
plt.show()
```
